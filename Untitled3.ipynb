```python
#!/usr/bin/env python
# coding: utf-8

import os
import pathlib
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from PIL import Image

# Set image dimensions and batch size
img_height, img_width = 180, 180
batch_size = 32

# Download and extract the dataset
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin=dataset_url, untar=True)
data_dir = pathlib.Path(data_dir)

# Handle nested directory structure if needed
if (data_dir / 'flower_photos').exists():
    data_dir = data_dir / 'flower_photos'

# Confirm dataset structure
print("Data directory:", data_dir)
print("Subdirectories:", list(data_dir.iterdir()))

# List count of images in each class
for class_folder in data_dir.glob('*'):
    if class_folder.is_dir():
        images = list(class_folder.glob('*'))
        print(f"{class_folder.name}: {len(images)} images")

# Pick and display a sample rose image
roses = list(data_dir.glob('roses/*'))
if not roses:
    raise ValueError("No rose images found. Check dataset structure.")
print("Sample image path:", roses[0])
Image.open(str(roses[0])).show()

# Load dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)

# Class names
class_names = train_ds.class_names
num_classes = len(class_names)
print("Classes:", class_names)

# Normalize and one-hot encode
train_ds = train_ds.map(lambda x, y: (x / 255.0, tf.one_hot(y, depth=num_classes)))
val_ds = val_ds.map(lambda x, y: (x / 255.0, tf.one_hot(y, depth=num_classes)))

# Prefetch for performance
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)

# Visualize sample images
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(6):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy())
        plt.title(class_names[tf.argmax(labels[i]).numpy()])
        plt.axis("off")
plt.show()

# Load pretrained ResNet50 model
base_model = tf.keras.applications.ResNet50(
    input_shape=(img_height, img_width, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)
base_model.trainable = False  # Freeze base model

# Create full model
model = Sequential([
    base_model,
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])

model.summary()

# Compile model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
epochs = 10
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)

# Plot accuracy
plt.figure()
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.4, 1])
plt.legend()
plt.grid()
plt.show()

# Plot loss
plt.figure()
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()

# Predict on a sample image
img_path = str(roses[0])
image = Image.open(img_path).resize((img_width, img_height))
image = np.array(image) / 255.0
image = np.expand_dims(image, axis=0)

# Make prediction
predictions = model.predict(image)
predicted_class = class_names[np.argmax(predictions)]
print(f"The predicted class is: {predicted_class}")
```
